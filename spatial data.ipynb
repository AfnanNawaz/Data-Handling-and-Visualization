```python
import folium

# Create a map centered around Bangalore
bangalore_map = folium.Map(location=[12.9716, 77.5946], zoom_start=12)

# Add markers for specific areas in Bangalore
places = [("M.G. Road", 12.9719, 77.6075),
          ("Indiranagar", 12.9784, 77.6408),
          ("Koramangala", 12.9352, 77.6245),
          ("Jayangar", 12.9295, 77.5824)]

for place in places:
    folium.Marker([place[1], place[2]], popup=place[0]).add_to(bangalore_map)

# Display the map
bangalore_map
```
